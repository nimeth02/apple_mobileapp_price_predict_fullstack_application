```python
import matplotlib.pyplot as plt

models = ['iPhone 11', 'iPhone 7', 'iPhone XS', 'iPhone X', 'iPhone 12 Pro Max',
          'iPhone 7 Plus', 'iPhone 6S', 'iPhone XS Max', 'iPhone 6', 'iPhone 8',
          'iPhone 12', 'iPhone 6S Plus', 'iPhone XR', 'iPhone 8 Plus', 'iPhone 4',
          'iPhone 12 Pro', 'iPhone 5S', 'iPhone 13 Pro', 'iPhone 11 Pro',
          'iPhone 12 mini', 'iPhone 13 Pro Max', 'iPhone 11 Pro Max', 'iPhone 13',
          'iPhone SE', 'iPhone 6 Plus', 'iPhone 5', 'iPhone SE 2', 'iPhone 4S',
          'iPhone 5C', 'iPhone 3GS', 'iPhone 13 mini', 'Other model']


fig, axs = plt.subplots(8,4, figsize=(30,40))


axs = axs.flatten()


for i, m in enumerate(models):
    cdf = df[df['Model'] == m]
    axs[i].scatter(cdf['USA'], cdf['prc'])
    axs[i].set_title(m)  
    axs[i].set_xlabel('USA')
    axs[i].set_ylabel('price')

# Hide any empty subplots
for i in range(len(models), len(axs)):
    axs[i].axis('off')

plt.tight_layout()
plt.show()
```
